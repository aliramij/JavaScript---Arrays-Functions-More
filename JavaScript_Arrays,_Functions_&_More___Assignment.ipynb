```python
# 2. Write a function checkEligibility that takes an age as a parameter
# and returns:
# ● "Eligible" if age is 18 or above
# ● "Not Eligible" if age is below 18
# Use an if-else statement.

'''
function checkEligibility(age){
    if(age == 18 || age>18){
        console.log("Eligible");
    }
    else if(age<18){
        console.log("Not Eligible");
    }
}
checkEligibility(17);

'''
```

```python
# 3. Write a function getDayName that takes a number (1-7) and returns the
# corresponding day of the week. Use a switch statement.

'''
function getDayName(dayNumber) {
  switch (dayNumber) {
    case 1:
      return "Monday";
    case 2:
      return "Tuesday";
    case 3:
      return "Wednesday";
    case 4:
      return "Thursday";
    case 5:
      return "Friday";
    case 6:
      return "Saturday";
    case 7:
      return "Sunday";
    default:
      return "Invalid Number Please try again";
  }
}
let result = getDayName(8);
console.log(result);

'''
```

```python
# 4.  Rewrite the checkEligibility function from Question 2 using the
# ternary operator.

'''
function checkEligibility(age){
    return age>18 || age == 18 ? "Eligible" : "Not Eligible";
}
console.log(checkEligibility(19));

'''
```

```python
# 8. Write a function greet that takes a name and a callback function. The greet function should call the callback with the name.

'''
function greet(name,callback){
    callback(name);
}
function call(name){
    console.log("Hello " + name);
}
greet("Sir",call);

'''
```
